```python
import pandas as pd
```

```python
df = pd.read_csv("csse_covid_19_daily_reports.csv.gz", index_col=0)
```

```python
df
```

```python
df['Country/Region'] = df['Country/Region'].replace("US", 'United States of America')
df['Country/Region'] = df['Country/Region'].replace("Mainland China", 'China')
```

```python
df.head()
```

```python
df['Active'] = df['Confirmed'] - df['Recovered']
```

```python
df.index = pd.DatetimeIndex(df['Last Update'])
```

```python
df.plot()
```

```python
df.to_csv('./enriched.csv')
```
