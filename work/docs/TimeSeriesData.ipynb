```python
from pathlib import Path
import pandas as pd

path = Path('/tmp/COVID-19-master/csse_covid_19_data/csse_covid_19_time_series/')
files = ['time_series_19-covid-Confirmed.csv', 'time_series_19-covid-Recovered.csv', 'time_series_19-covid-Deaths.csv']
df = pd.DataFrame(columns=['Province/State', 'Country/Region', 'Lat', 'Long', 'Date'])
for f in files:
    other = pd.read_csv(
        path.joinpath(f),
    )
    other = other.melt(id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], var_name='Date', value_name=(path / f).stem)
    df = pd.merge(df, other, how='outer')

df.to_csv('../samples/time_series_19-covid.csv', index_label='ix')
```
