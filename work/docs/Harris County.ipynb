```python
def trim(obj):
    try:
        return obj.strip()
    except ValueError:
        return string

def split_sex(string):
    return trim(string.split('-')[-1])
```

```python
df, *rest = pd.read_html(url, header=0, attrs={'id': 'covidtable'})
```

```python
df = df.applymap(trim)
```

```python
df['Sex'] = df.Sex.apply(split_sex)
```
